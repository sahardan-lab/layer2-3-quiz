```python
import google.generativeai as genai
from IPython.display import display, HTML

# 1. הגדרת ה-API Key שלך
genai.configure(api_key="AIzaSyCwZrLROHZJ7CJjmM1uJfeZxxiTkY74gys")

# 2. מנגנון בחירת מודל אוטומטי למניעת שגיאת 404
def get_working_model():
    try:
        # מחפש מודל שתומך ביצירת תוכן (בדרך כלל gemini-1.5-flash או gemini-pro)
        for m in genai.list_models():
            if 'generateContent' in m.supported_generation_methods:
                return genai.GenerativeModel(m.name)
    except:
        return genai.GenerativeModel('gemini-1.5-flash') # ברירת מחדל

model = get_working_model()
chat = model.start_chat(history=[])

# 3. פונקציית הבועות המטורפות (מעוצב ומיושר לימין)
def show_bubble(text, role='bot'):
    if role == 'bot':
        bg = "#e8f0fe"; border = "2px solid #4285f4"; label = "🤖 NetCoach"
    else:
        bg = "#f1f3f4"; border = "1px solid #dadce0"; label = "👤 אתה"

    html = f"""
    <div style="direction: rtl; text-align: right; margin: 8px 0;">
        <span style="font-family: Arial; font-size: 11px; color: #70757a; margin-right: 5px;">{label}</span><br>
        <div style="background-color: {bg}; border: {border}; border-radius: 18px;
                    padding: 10px 15px; display: inline-block; max-width: 75%;
                    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
                    font-size: 15px; line-height: 1.4; box-shadow: 1px 1px 3px rgba(0,0,0,0.1);">
            {text.replace('\n', '<br>')}
        </div>
    </div>
    """
    display(HTML(html))

# 4. הנחיות הבוחן (קצר, קולע, אמריקאי)
system_context = """
אתה NetCoach, בוחן רשתות ממוקד.
חוקים:
1. שאל רק שאלות אמריקאיות (א, ב, ג, ד) על: NIC, MAC, IP, Routing.
2. משוב קצר: אם צדק - "כל הכבוד!". אם טעה - הסבר של 2 שורות בלבד.
3. תמיד תסיים בשאלה חדשה.
4. אל תכתוב הסברים ארוכים!
"""

show_bubble("אחונמסקי! המערכת תוקנה. מוכן למבחן אמריקאי קצר על שכבות 1-3? כתוב 'התחל'.", 'bot')

while True:
    user_input = input("התשובה שלך: ")
    if user_input.lower() in ['exit', 'יציאה']: break

    show_bubble(user_input, 'user')

    try:
        prompt = f"{system_context}\nתשובת המשתמש: {user_input}\nתן משוב ושאלה חדשה."
        response = chat.send_message(prompt)
        show_bubble(response.text, 'bot')
    except Exception as e:
        show_bubble(f"אופס, יש תקלה טכנית קטנה. נסה שוב! (שגיאה: {str(e)[:50]})", 'bot')
```
