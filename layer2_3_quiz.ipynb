```python
import google.generativeai as genai
import textwrap

# הגדרת ה-API
genai.configure(api_key="AIzaSyBYVphkLZGMQoZdbSsdeRhU2rg6yfL2Ffk")
model = genai.GenerativeModel('models/gemini-2.5-flash')

# הנחיות לסוכן - התמקדות באינפוגרפיקות ועיצוב טקסט
system_context = """
אתה מנהל משחק לימודי על רשתות תקשורת.
השאלות שלך חייבות להתבסס על:
1. שכבה 1 (פיזית): סיגנלים, כבלי נחושת, סיבים אופטיים ו-Wi-Fi.
2. שכבה 2 (Data-Link): כרטיס רשת (NIC), כתובות MAC, מסגרות (Frames) ומתגים (Switches).
3. שכבה 3 (Network): כתובות IP ונתבים (Routers).

חוק חשוב:
אל תכתוב שורות ארוכות! בכל פעם שאתה כותב הסבר או שאלה, תרד שורה אחרי כל 10-12 מילים כדי שהטקסט לא יימרח לרוחב המסך.
"""

def print_wrapped(text):
    # פונקציה שדואגת שההדפסה ב-Colab תהיה צרה וקריאה
    lines = text.split('\n')
    for line in lines:
        print(textwrap.fill(line, width=60))

chat = model.start_chat(history=[])

print("--- ברוך הבא למבוך הרשתות של אחונם! ---")
print("הסוכן בוחן אותך על שכבות 1-3. מוכן?")

while True:
    user_input = input("\nהתשובה/פעולה שלך: ")
    if user_input.lower() in ['יציאה', 'exit']:
        break

    # שליחה לסוכן
    response = chat.send_message(f"{system_context} \n השחקן אמר: {user_input}")

    print("\n" + "-"*30)
    print_wrapped(response.text)
    print("-"*30)
```
